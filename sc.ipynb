```python
from scapy import all as scapy

packets = []


class TCPAnalyzer:  
    def __init__(self, packet):
        self.packet = packet
        self.modify()

    def modify(self):
        # Check if it's a TCP packet
        if self.packet.haslayer(scapy.TCP):
            tcp_layer = self.packet[scapy.TCP]
            
            # Modify the TCP flags (e.g., set only the ACK flag)
            tcp_layer.flags = "F"  # 'A' is for ACK, modify this based on your requirement
            
            # Send the modified packet
            scapy.send(self.packet)
            packets.append(self.packet)
            print(f"Sent modified packet with flags: {tcp_layer.flags}")

    def __repr__(self):
        return self.packet.__str__()

    def __str__(self):
        return self.packet.__str__()


# Sniff TCP packets on the specified interface and apply TCPAnalyzer
scapy.sniff(iface="enx04bf1b9ed863", filter="tcp", prn=TCPAnalyzer)

# Print captured packets
print(packets)
```
